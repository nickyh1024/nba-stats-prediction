```python
import pandas as pd 

df = pd.read_csv('../data/2019-2024_combined_stats.csv')

print(df.head())
print(df.columns)
df.head()
```

```python
# drop columns that aren't needed
df = df.drop(columns=["Rk", "Awards"], errors='ignore')
df.head()
```

```python
# handle missing values for percentages
for col in ["FG%", "3P%", "2P%", "eFG%", "FT%"]:
    if col in df.columns:
        df[col] = df[col].fillna(0)

# handle missing values for everything else
df = df.fillna(0)

df.isnull().sum().sum()
```

```python
df["Player"].value_counts().head(10)
```

```python
df.info()
df.describe().T.head()
```

```python
# data cleaning

# there is no season column, but we can use the age column to sort by season
df = df.sort_values(by=["Player", "Age"]).reset_index(drop=True)

# Create next season targets
# only do points, rebounds, assists, steals, blocks, turnovers, 3P, FG%, FT%
target_cols = ["PTS", "TRB", "AST", "STL", "BLK", "TOV", "3P", "FG%", "FT%"]

for col in target_cols:
    df[f"{col}_next"] = df.groupby("Player")[col].shift(-1)

# drop rows where the next season's stats are missing (so no following season)
next_cols = [f"{c}_next" for c in target_cols]
df_model = df.dropna(subset=next_cols).copy()

df_model.shape, df_model[next_cols].head()

# inputs (current season stats)
feature_cols = [
    "Age", "G", "MP", "FGA", "3PA", "FTA", "FG", "3P", "FT", "TRB", "AST", "STL", "BLK",
    "TOV", "PTS", "FG%", "FT%",
]

# keep columns that actually exist
feature_cols = [c for c in feature_cols if c in df_model.columns]

X = df_model[feature_cols].copy()
Y = df_model[[f"{c}_next" for c in target_cols]].copy()

X.head(), Y.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.multioutput import MultiOutputRegressor

# train/test split
# use last season (23-24) as test set, earlier seasons as train 
# all but oldest seasons
train = df_model[df_model["Age"] < df_model["Age"].max()]
# last age is last season which is test
test = df_model[df_model["Age"] == df_model["Age"].max()]

X_train, Y_train = train[feature_cols], train[[f"{c}_next" for c in target_cols]]
X_test, Y_test = test[feature_cols], test[[f"{c}_next" for c in target_cols]]

# choose model
rf = RandomForestRegressor(n_estimators=200, random_state=42)

# wrap for multi-output
model = MultiOutputRegressor(rf)

# train
model.fit(X_train, Y_train)
```

```python
from sklearn.metrics import mean_absolute_error

# evaluating model performance
Y_pred = model.predict(X_test)

# evaluate each stat separately
mae_scores = {}
for i, col in enumerate(Y_test.columns):
    mae = mean_absolute_error(Y_test.iloc[:, i], Y_pred[:, i])
    mae_scores[col] = mae

mae_scores
```

```python
# compare actual vs predicted for a few players to see how well the model is doing
results = X_test.copy()
results["Player"] = test["Player"].values
for i, col in enumerate(Y_test.columns):
    results[f"{col}_actual"] = Y_test.iloc[:, i].values
    results[f"{col}_pred"]= Y_pred[:, i]

print(results)
```

```python
# model only returns oldest player so need to create a season column
# we can do this by sorting each player's seasons by age and then assigning the year
# first we need to further clean the data, players show up multiple times in one season if they got traded

import numpy as np
import pandas as pd

df = df.copy()

df = df.sort_values(["Player", "Age"]).reset_index(drop=True)

# helpers
AGG_TEAM_TOKENS = {"TOT", "2TM", "3TM", "4TM"}

# columns we will sum up
sum_cols = ["G", "GS"]
sum_cols = [c for c in sum_cols if c in df.columns]

# numeric columns to games-weight-average (most per-game stats)
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()
exclude_from_weight_avg = set(sum_cols + ["Age"])
weight_avg_cols = [c for c in num_cols if c not in exclude_from_weight_avg]

def collapse_player_age(group: pd.DataFrame) -> pd.DataFrame:
    """
    Collapse rows for a single (Player, Age) to one row. Returns a single-row DataFrame.
    """
    # if there's a combined row return it
    if "Team" in group.columns and group["Team"].isin(AGG_TEAM_TOKENS).any():
        row = group[group["Team"].isin(AGG_TEAM_TOKENS)].iloc[0].copy()
        return row.to_frame().T
    
    # compute games-weighted averages
    out = group.iloc[0].copy()

    # sum games and games started
    for c in sum_cols:
        out[c] = group[c].sum() if c in group.columns else np.nan

    # weight = games played ()

    # Weight = games played (fallback to equal weights if G missing or zeros)
    w = group["G"] if "G" in group.columns else pd.Series(1.0, index=group.index)
    # Avoid all-zero division
    w = w.replace(0, np.nan)
    w = w.fillna(w.mean() if np.isfinite(w.mean()) else 1.0)

    # Games-weighted average for numeric stat columns
    for c in weight_avg_cols:
        # Keep Age from first row (already handled in exclude set)
        if c in group.columns:
            out[c] = np.average(group[c].values, weights=w.values)

    # Keep a representative team label (optional)
    if "Team" in group.columns:
        out["Team"] = "MULTI"  # or group.iloc[group["G"].idxmax()]["Team"]

    return out.to_frame().T

# 3) Apply collapse on (Player, Age)
collapsed = (
    df.groupby(["Player", "Age"], as_index=False, group_keys=False)
    .apply(collapse_player_age)
    .reset_index(drop=True)
)

collapsed.shape, collapsed.head()
```

```python
# create a season column

collapsed = collapsed.sort_values(["Player", "Age"]).reset_index(drop=True)
collapsed["Season"] = collapsed.groupby("Player").cumcount() + 2019

# check each player + season is unique
assert not collapsed.duplicated(["Player", "Season"]).any()

df = collapsed.copy()

target_cols = ["PTS", "TRB", "AST", "STL", "BLK", "TOV", "3P", "FG%", "FT%"]

for col in target_cols:
    df[f"{col}_next"] = df.groupby("Player")[col].shift(-1)

next_cols = [f"{c}_next" for c in target_cols]
df_model = df.dropna(subset=next_cols).copy()

# Features â€” tweak as you like
feature_cols = [
    "Age", "G", "MP",
    "FGA", "3PA", "FTA",
    "FG", "3P", "FT",
    "TRB", "AST", "STL", "BLK", "TOV", "PTS",
    "FG%", "FT%",
]
feature_cols = [c for c in feature_cols if c in df_model.columns]

test_season = df_model["Season"].max()
train = df_model[df_model["Season"] < test_season]
test = df_model[df_model["Season"] == test_season]

X_train, Y_train = train[feature_cols], train[next_cols]
X_test,  Y_test  = test[feature_cols],  test[next_cols]

print(f"Train: {X_train.shape}, Test: {X_test.shape}    (test_season={test_season})")

rf = RandomForestRegressor(
    n_estimators=300,
    max_depth=None,      # let trees grow; you can tune later
    random_state=42,
    n_jobs=-1
)

model = MultiOutputRegressor(rf)
model.fit(X_train, Y_train)

Y_pred = model.predict(X_test)

mae_scores = {}
for i, col in enumerate(Y_test.columns):
    mae_scores[col] = round(mean_absolute_error(Y_test.iloc[:, i], Y_pred[:, i]), 3)

pd.Series(mae_scores).sort_index()

# Helpful to keep player/season identifiers
results = test[["Player", "Season"]].copy()

# Add predictions + actuals
for i, col in enumerate(Y_test.columns):
    results[f"{col}_actual"] = Y_test.iloc[:, i].values
    results[f"{col}_pred"]   = Y_pred[:, i]

# Sort by biggest absolute PTS error as an example
results_with_err = results.copy()
results_with_err["PTS_abs_err"] = (results_with_err["PTS_next_actual"] - results_with_err["PTS_next_pred"]).abs()

results_with_err.sort_values("PTS_abs_err", ascending=False).head(15)

# Average MAE across all 9 targets
avg_mae = np.mean(list(mae_scores.values()))
print("Average MAE across targets:", round(avg_mae, 3))

# Show MAE nicely formatted
mae_df = pd.DataFrame({"MAE": pd.Series(mae_scores)}).sort_values("MAE")
mae_df

# Map index -> target name
target_by_idx = list(Y_train.columns)
target_idx = target_by_idx.index("PTS_next")  # choose a target to inspect
rf_for_pts = model.estimators_[target_idx]

fi = pd.Series(rf_for_pts.feature_importances_, index=feature_cols).sort_values(ascending=False)
fi.head(15)
```
