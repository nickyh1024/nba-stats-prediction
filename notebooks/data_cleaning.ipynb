```python
import pandas as pd 

df = pd.read_csv('../data/2019-2024_combined_stats.csv')

print(df.head())
print(df.columns)
df.head()
```

```python
# drop columns that aren't needed
df = df.drop(columns=["Rk", "Awards"], errors='ignore')
df.head()
```

```python
# handle missing values for percentages
for col in ["FG%", "3P%", "2P%", "eFG%", "FT%"]:
    if col in df.columns:
        df[col] = df[col].fillna(0)

# handle missing values for everything else
df = df.fillna(0)

df.isnull().sum().sum()
```

```python
df["Player"].value_counts().head(10)
```

```python
df.info()
df.describe().T.head()
```

```python
# data cleaning

# there is no season column, but we can use the age column to sort by season
df = df.sort_values(by=["Player", "Age"]).reset_index(drop=True)

# Create next season targets
# only do points, rebounds, assists, steals, blocks, turnovers, 3P, FG%, FT%
target_cols = ["PTS", "TRB", "AST", "STL", "BLK", "TOV", "3P", "FG%", "FT%"]

for col in target_cols:
    df[f"{col}_next"] = df.groupby("Player")[col].shift(-1)

# drop rows where the next season's stats are missing (so no following season)
next_cols = [f"{c}_next" for c in target_cols]
df_model = df.dropna(subset=next_cols).copy()

df_model.shape, df_model[next_cols].head()

# inputs (current season stats)
feature_cols = [
    "Age", "G", "MP", "FGA", "3PA", "FTA", "FG", "3P", "FT", "TRB", "AST", "STL", "BLK",
    "TOV", "PTS", "FG%", "FT%",
]

# keep columns that actually exist
feature_cols = [c for c in feature_cols if c in df_model.columns]

X = df_model[feature_cols].copy()
Y = df_model[[f"{c}_next" for c in target_cols]].copy()

X.head(), Y.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.multioutput import MultiOutputRegressor

# train/test split
# use last season (23-24) as test set, earlier seasons as train 
# all but oldest seasons
train = df_model[df_model["Age"] < df_model["Age"].max()]
# last age is last season which is test
test = df_model[df_model["Age"] == df_model["Age"].max()]

X_train, Y_train = train[feature_cols], train[[f"{c}_next" for c in target_cols]]
X_test, Y_test = test[feature_cols], test[[f"{c}_next" for c in target_cols]]

# choose model
rf = RandomForestRegressor(n_estimators=200, random_state=42)

# wrap for multi-output
model = MultiOutputRegressor(rf)

# train
model.fit(X_train, Y_train)
```
